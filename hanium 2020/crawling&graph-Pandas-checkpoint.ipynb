```python
#라이브러리

# 라이브러리를 로드합니다.
from bs4 import BeautifulSoup
import requests
import re
import sys
import pprint

#word cloud에 필요한 lib
from konlpy.tag import Kkma
from collections import Counter
import pytagcloud
import webbrowser


#그래프 그리기
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#댓글 크롤링

# 댓글을 달 빈 리스트를 생성합니다.
CommentList = []
PKList = []
RegTList =[]
SympathList=[]
AntipathList=[]
# 네이버 뉴스 url을 입력합니다.
url = "https://news.naver.com/main/read.nhn?m_view=1&includeAllCount=true&mode=LSD&mid=shm&sid1=100&oid=422&aid=0000430957"

oid = url.split("oid=")[1].split("&")[0] #422
aid = url.split("aid=")[1] #0000430957
page = 1
header = {
    "User-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36",
    "referer": url,

}
while True:
    line = [] #안쪽 리스트로 사용할 빈 리스트 
    c_url = "https://apis.naver.com/commentBox/cbox/web_neo_list_jsonp.json?ticket=news&templateId=default_society&pool=cbox5&_callback=jQuery1707138182064460843_1523512042464&lang=ko&country=&objectId=news" + oid + "%2C" + aid + "&categoryId=&pageSize=20&indexSize=10&groupId=&listType=OBJECT&pageType=more&page=" + str(
        page) + "&refresh=false&sort=FAVORITE"
    # 파싱하는 단계입니다.
    r = requests.get(c_url, headers=header)
    cont = BeautifulSoup(r.content, "html.parser")
    total_comm = str(cont).split('comment":')[1].split(",")[0]
    
    c_match = re.findall('"contents":([^\*]*),"userIdNo"', str(cont))
    pk = re.findall('"commentNo":([^\*]*),"parentCommentNo"', str(cont))
    reg_time = re.findall('"regTime":([^\*]*),"regTimeGmt"', str(cont))
    sympathy = re.findall('"sympathyCount":([^\*]*),"antipathyCount"', str(cont))
    antipathy= re.findall('"antipathyCount":([^\*]*),"userBlind"', str(cont))
    
    #각 col에 해당하는 list에 추가
    CommentList.append(c_match)#댓글 내용 리스트에 추가
    PKList.append(pk)#CommentNo(pk) 리스트에 추가
    RegTList.append(reg_time) #등록시간 리스트에 추가
    SympathList.append(sympathy) #공감수 
    AntipathList.append(antipathy) #비공감수
    
    # 한번에 댓글이 20개씩 보이기 때문에 한 페이지씩 몽땅 댓글을 긁어 옵니다.
    if int(total_comm) <= ((page) * 20):
        break
    else:
        page += 1
   
 # 여러 리스트들을 하나로 묶어 주는 함수입니다.
def flatten(l):
    flatList = []
    for elem in l:
        # if an element of a list is a list
        # iterate over this list and add elements to flatList
        if type(elem) == list:
            for e in elem:
                flatList.append(e)
        else:
            flatList.append(elem)
    return flatList

def makeRow(l1,l2,l3,l4,l5): #[commnetNo, 댓글내용, 작성시간, 공감수, 비공감수] elem이 포함된 리스트를 만든다.
    fl1=flatten(l1)
    fl2=flatten(l2)
    fl3=flatten(l3)
    fl4=flatten(l4)
    fl5=flatten(l5)
    RowList = []
    list_len = min(len(fl1),len(fl2),len(fl3),len(fl4),len(fl5))
    for i in range(list_len):
        temp = []
        temp.append(fl1[i])
        temp.append(fl2[i])
        temp.append(fl3[i])
        temp.append(fl4[i])
        temp.append(fl5[i])
        RowList.append(temp)
    return RowList
    

# 리스트 결과입니다.
rows = makeRow(PKList,CommentList,RegTList,SympathList,AntipathList)
for i in rows : #타입변환
    i[2] = i[2].split('"')[1].split(":")[0] #시단위 까지만 가져옴
    #i[2]=maya.parse(i[2][1]).datetime() #등록시간 type을 <class 'datetime.datetime'>로 변경
   #댓글번호, 공감, 비공감 수 int타입으로 변환
    i[0] = int(i[0])
    i[3] = int(i[3]) 
    i[4] = int(i[4])
comment_list = pd.DataFrame(rows,columns=['CommentNo','Content','RegTime','SympathyCount','antipathyCount'])
comment_list
```

```python
#분석 결과 시각화 
#1.워드클라우드
kkma = Kkma()
List = flatten(CommentList) #댓글 내용만 DB에서 불러온다.(DB연동 후 수정 필요)

ex_nouns=[] #단어저장
for sen in List :
    for noun in kkma.nouns(sen) : #sentence -> noun
        ex_nouns.append(noun)

#명사 전처리 필요
pre_noun =[] #한음절 단어 삭제
for i in ex_nouns :
    if len(i) > 1 :
        pre_noun.append(i)



# top30
counter = Counter(pre_noun)
top30 = counter.most_common(n=50)#빈도수가 많은 30개 추출

word_count_list = pytagcloud.make_tags(top30, maxsize=90)
# maxsize : 최대 글자크기
print(word_count_list)

pytagcloud.create_tag_image(word_count_list,'newsCrawling.jpg', size=(900, 600), fontname='korean', rectangular=False) #크기 조절필요

 
webbrowser.open('newsCrawling.jpg')
```

```python
#2.시간별 댓글 수 그래프 
comment_list['RegTime'] = pd.to_datetime(comment_list['RegTime'], format='%Y-%m-%d', errors='raise')
time_group = comment_list.groupby(['RegTime']).count() #시간대 별로 그룹을 만들어 count한다.
temp = time_group.index.tolist()
xtext =[]
for i in temp : #x축을 n일 n시 형식으로 변환
    i = str(i)
    i = i.split("-")[2].split(" ")[0]+"일"+i.split(" ")[1].split(":")[0] +"시" 
    xtext.append(i)
    
plt.xticks(time_group.index,xtext,fontsize=10,rotation=90)
plt.rc('figure',figsize=(15,5)) # 그래픽 크기 설정
plt.rc('font',size=16)
plt.title('시간별 댓글수')
xtext = time_group.index.tolist()
#plt.style.use('ggplot') #격자무늬
time_group['Content'].plot(color = 'green')
plt.show()
time_group
```

```python
#3.공감, 비공감수 분석

# 3-1.공감의 수가 더 많은 댓글과 비공감 수가 더 많은 댓글의 수 비교(bar graph)
comment_list['gap']=comment_list['SympathyCount']-comment_list['antipathyCount'] 
#공감 - 비공감이 양수이면 공감이 더 많은 댓글 음수이면 비공감이 더 많은 댓글
pos_count = len(comment_list[comment_list['gap']>0]) #공감의 수가 더 많은 댓글의 수
neg_count = len(comment_list[comment_list['gap']<0]) #비공감의 수가 더 많은 댓글의 수
pos_neg = [pos_count,neg_count]
if neg_count+pos_count >0 :#모든 댓글에 공감, 비공감이 없을 경우 그래프를 표시하지 않는다.
    label1 = ['공감이 많은 댓글','비공감이 많은 댓글']
    plt.rc('font',family='Malgun Gothic')
    plt.title('공감/비공감 댓글 수')
    plt.rc('figure',figsize=(4,4))
    plt.bar(label1,pos_neg,width=0.5,color = ['tomato','cornflowerblue'])
    plt.show()
#공감,비공감 댓글 top3 추출
sort_pos_top3 = comment_list.sort_values(by="gap", ascending=False).head(3)
sort_neg_top3 = comment_list.sort_values(by="gap", ascending=True).head(3)

print('공감을 가장 많이 받은 댓글 TOP3 ')
print (sort_pos_top3['Content'])
print('   ')
print('비공감을 가장 많이 받은 댓글 TOP3 ')
print (sort_neg_top3['Content'])

    
    
#comment_list[comment_list['gap']>0].max()
```

```python
#3-2.공감 비공감의 수가 비슷한 댓글(논쟁)

#공감+비공감의 수가 상위 10%만 추출 
comment_list['pathy_sum'] = comment_list['SympathyCount']+comment_list['antipathyCount']
battle_list = comment_list.sort_values(by="pathy_sum", ascending=False).head(int(len(comment_list)*0.1))
battle_list['abs_gap'] = abs(battle_list['gap']) #비교를 위해 gap을 절대값으로 변환
#공감과 비공감 퍼센트의 차이를 구한다. =>댓글의 수가 다르기 때문에 퍼센트로 계산
battle_list['battle'] = (battle_list['abs_gap']/battle_list['pathy_sum'])*100
battle_sort = battle_list.sort_values(by='battle',ascending=True)

#가장 논쟁이 많은 댓글의 공감, 비공감 수를 저장할 list
battle_list1=[] 

#각 리스트마다 가장 논쟁이 많은 댓글의 [공감수,비공감수,내용] 저장
# 공감수만 리스트로 출력
battle_list1 = battle_sort.SympathyCount.head(1).tolist() 

#비공감수 추가
battle_list1.append(battle_sort.antipathyCount.head(1).tolist()[0]) 

#댓글 내용 추가 
battle_list1.append(battle_sort.Content.head(1).tolist()[0])

#그래프 그리기 - 그래프 모양 다른 것도 생각해보기
label1 = ['공감수','비공감수']
plt.axis('equal') #동그란 원으로 표현
plt.title('논쟁 중인 댓글 1위')
plt.pie(battle_list1[:2],labels=label1,autopct='%.1f%%',colors=['tomato','cornflowerblue']) # 그래프에는 공감, 비공감만 나타낸다.
plt.show()
print(battle_list1[2])
print(" ")
```
