```python
#상호,업주명 검색을 통한 로그인 후 종합소득세 화면으로 넘어가는 매크로
#serach id
while True :
    name = input('이름 : ')
    if name in 'store_name1' or name in 'owner_name1' : #가상의 업체, 이름임
        my_id = 'id1'
        break
    elif name in 'store_name2' or name in 'owner_name2' :
        my_id = 'id2'
        break
    elif name in 'store_name3' or name in 'owner_name3':
        my_id = 'id3'
        break
    
    elif name in 'store_name4' or name in 'owner_name4' :
        my_id = 'id4'
        break
        
        
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.alert import Alert
driver = webdriver.Chrome()
import time
import pyautogui as pag
import time        
driver.maximize_window() #화면크기 최대

#3초 기다린 후 홈페이지 접속
driver.implicitly_wait(3)
driver.get('https://www.hometax.go.kr/websquare/websquare.wq?w2xPath=/ui/comm/a/b/UTXPPABA01.xml&redirectMenuId=2301000000&w2xHome=/ui/pp/&w2xDocumentRoot=')
driver.implicitly_wait(10)
#로그인 화면

elem = driver.find_element_by_id('iptUserId')
elem.clear()
#초기화
elem.send_keys(my_id)

my_pw ='temp_pw' #가상의 비밀번호임
pw=driver.find_element_by_id('iptUserPw')
pw.clear()
pw.send_keys(my_pw)

driver.find_element_by_id('trigger46').click()
time.sleep(3)

# iframe 전환 : 공인인증서 로그인 화면
iframe = driver.find_element_by_css_selector('#dscert')
driver.switch_to.frame(iframe)

# 공인인증서 선택
driver.find_element_by_xpath('//*[@title="공인인증서명"]').click()
driver.implicitly_wait(3)

# 패스워드 입력
passwd = 'temp_pw'#가상의 비밀번호
driver.find_element_by_id('input_cert_pw').send_keys(passwd)
driver.implicitly_wait(3)
    
    
# 확인버튼 클릭
driver.find_element_by_id('btn_confirm_iframe').click()
driver.implicitly_wait(3)

# iframe원복
driver.switch_to.default_content()
 

iframe = driver.find_element_by_css_selector('#txppIframe')
driver.switch_to.frame(iframe)
driver.implicitly_wait(10)
pag.typewrite('enter')





#파일변환 신고


driver.find_element_by_xpath('//*[@id="anchor17"]').click()
driver.implicitly_wait(3)

#팝업창 해결
try:
    da = driver.switch_to.alert
    da.accept()
    
except : 
    time.sleep(0.1) #팝업창이 없을 경우 넘어감



    
```
